```python
#import neccessary module
import sqlite3
import time
import datetime
import random
```

```python
#connect sqlite3 module with the database
conn = sqlite3.connect('enron.db')
```

```python
#to load the data into the database 
cur = conn.cursor()
```

```python
#create employee list table
sql = """
CREATE TABLE IF NOT EXISTS users(
   Email_id TEXT PRIMARY KEY,
   eid INT
   firstName TEXT,
   lastName TEXT,
   Email2 REAL,
   Email3 REAL,
   Email4 REAL,
   folder BLOB,
   status TEXT);
"""

#Execute query and commit transaction
cur.execute(sql)
conn.commit()
```

```python
#connecting the employee list and printing the tuples
rows = cur.execute("SELECT Email_id, eid, firstname, Email2, Email3, Email4,folder, status FROM employeelist").fetchall()
print(rows)
```

```python
#selecting all users from the employee lis
sql = '''
SELECT * FROM employeelist;
'''

#Execute query and commit transaction  and printing the list
rows = cur.execute(sql)
for row in rows:
  print(row)
```

```python
#importind a data frame which can load data in table
import pandas as pd

sql='''
SELECT * from employeelist;
'''
# Load the data into a DataFrame
users_df = pd.read_sql_query(sql,conn) 
users_df
```

```python
#create a table of messages
sql = """
CREATE TABLE IF NOT EXISTS message(
   mid INT PRIMARY KEY,
   date REAL,
   subject TEXT,
   body TEXT,
   folder TEXT);
"""

#Execute query and commit transaction
cur.execute(sql)
conn.commit()
```

```python
#select date and message count from the messahes dataframe
sql = '''
SELECT date(date) AS date_sent, count(mid) AS message_count
    FROM message
    GROUP BY date_sent
    ORDER BY date_sent
'''

#Execute query and commit transaction then print the data
rows = cur.execute(sql)
for row in rows:
  print(row)
```

```python
#import pandas and create table for date sent and message count
import pandas as pd

sql='''
SELECT date(date) AS date_sent, count(mid) AS message_count
    FROM message
    GROUP BY date_sent
    ORDER BY date_sent
'''
# Load the data into a DataFrame and print table
message_df = pd.read_sql_query(sql,conn) 
message_df
```

```python
#import module to plot the graph of date sent against message count and print the graph
import matplotlib.pyplot as plt
message_df.plot(x = 'date_sent', y = 'message_count', figsize = (12,9))
plt.show()
```

```python
#select data message sender and mesage count the print the table in ascending order
sql='''
SELECT sender AS sender, count(mid) AS message_count
    FROM message
    GROUP BY sender
    ORDER BY message_count
'''

#Execute query and commit transaction
message_df = pd.read_sql_query(sql,conn) 
message_df
```

```python
#Select bettom ten senders from the table
message_df.tail(10)
```

```python
#Plot a bar graph entities in the table
message_df.tail(10).plot(x = 'sender', y = 'message_count', kind = 'bar')
plt.show()
```
